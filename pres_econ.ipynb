```python
import pandas as pd
import json
```

```python
dataFile_US_Econ = "Resources/us_econ_growth.csv"
econ_growth_df = pd.read_csv(dataFile_US_Econ)
econ_growth_df = econ_growth_df.set_index("Year")
econ_growth_df = econ_growth_df.rename(columns={"Value": "annual_gnp_growth_pct"})
econ_growth_df.head()
```

```python
cig_file = "Resources/sales-of-cigarettes-per-adult-per-day.csv"
cig_df = pd.read_csv(cig_file)
cig_df = cig_df.set_index("Year")
cig_df.head()
```

```python
with open('Resources/presidents_debt.json') as json_data:
    cig_data = json.load(json_data)
    #print(cig_data)
    
print(json.dumps(cig_data, indent=4, sort_keys=True))
```

```python
year_data = cig_data["data"]["by_year"]
year_data = pd.DataFrame(year_data)
year_data = year_data.rename(columns={"year": "Year"})
year_data.head()
```

```python
pres_year = year_data.set_index("Year")
pres_year.head()
```

```python
pres_econ = pd.merge(pres_year, econ_growth_df, on='Year')
pres_econ = pres_econ[["president", "party", "annual_gnp_growth_pct", "annual_debt_increase_pct"]]
pres_econ.head()
```

```python
ufo_file = "Resources/ufo.csv"
ufo_df = pd.read_csv(ufo_file)
ufo_df.head()
```

```python
yeartime = ufo_df['datetime'].str.split('/', n=3, expand = True) 
year = yeartime[2].str.split(' ', n=0, expand = True) 
year = year[0]
year = year.astype(int)
year.head()
```

```python
ufo_df['Year'] = year
ufo_df = ufo_df[ufo_df['country'] == 'us']
ufo_df_group = ufo_df.groupby(['Year']).agg(['count'])
ufo_df_group.head()
```

```python
ufo_df_group = ufo_df_group['datetime']
```

```python
ufo_df_group = ufo_df_group.rename(columns={'count':'ufo_sightings'})
ufo_df_group = pd.DataFrame(ufo_df_group)
ufo_df_group
```

```python
pres_econ_ufo = pd.merge(pres_econ, ufo_df_group, on='Year')
pres_econ_ufo
```
